```python
#Write numpy array as NetCDF
ncout = netCDF4.Dataset('RACMO2_accum_2.nc','w','NETCDF4_classic'); # using netCDF3 for output format 
ncout.createDimension('time',np.shape(PCP)[0])
ncout.createDimension('rlon',np.shape(PCP)[1])
ncout.createDimension('rlat',np.shape(PCP)[2])

timevar = ncout.createVariable('time','float64',('time')); timevar.setncattr('units','days since 1950-01-01 00:00:00.0'); timevar[:]=ncdf.variables['time'][:];

rlonvar = ncout.createVariable('rlon','float32',('rlon')); rlonvar[:] = ncdf.variables['rlon'][:];
lonvar = ncout.createVariable('lon','float32',('rlon','rlat')); lonvar[:] = ncdf.variables['lon'][:].T;

rlatvar = ncout.createVariable('rlat','float32',('rlat')); rlatvar[:] = ncdf.variables['rlat'][:];
latvar = ncout.createVariable('lat','float32',('rlon','rlat')); latvar[:] = ncdf.variables['lat'][:].T;

pcp_Accum = ncout.createVariable('pcp_Accum','float32',('time','rlon','rlat')); pcp_Accum.setncattr('units','kg m-2 s-1'); pcp_Accum[:] = PCP;

ncout.close();
```
